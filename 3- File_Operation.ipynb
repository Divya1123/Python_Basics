### File Operations

**Read full file or line by line**

```python
with open('file_example.txt', 'r') as file:
    content = file.read()   # read entire content of file
    print(type(content))
    print(content)
```

```python
with open('file_example.txt', 'r') as file:
    for line in file:
        print(line.strip())         # strip() remove any leading and trailing charcter(like /n, tab, spaces) from string
```

```python
with open('file_example.txt', 'a') as file:
    file.write("\n Hello World!")
```

```python
with open('file_example.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

**Writing multiple lines**

```python
lines = ['\nFirst line\n', 'Second line\n']

with open('file_example.txt', 'a') as file:
    file.writelines(lines)
```

```python
with open('file_example.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

**Read context from source text file and write to another text file**

```python
with open('file_example.txt', 'r') as source_file:
    content = source_file.read()

with open('target.txt', 'w') as destin_file:            # if not exits - it will create file
    destin_file.write(content)
```

'w+' file mode in python is used to open a file for both reading and writing. If file do not exist, it will be created. If file exist content will be overwritten.
And then content will be read. But cursor at end. To move cursor to beginning, we use seek(0)

```python
with open('target.txt', 'w+') as file:            # if not exits - it will create file
    file.write("Hello Divya\n")
    file.write("Hello World")

    # move cursor to the beginning - 0th index
    file.seek(0)

    # read content of file
    print(file.read())
```
