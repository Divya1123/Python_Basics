### Iterators<br>
Iterator is an object that represent a stream of data and allow to traverse through elements if collection(tuple, list etc) one at a time.<br>
Follows lazy loading and saves memory.

```python
my_list = [1,2,3,4]
print(type(my_list))
print(my_list)
```

```python
iterator = iter(my_list)
print(type(iterator))
print(iterator)
```

**next() - will give next element of iterator, but when no next element left, will give error. To handle this, will use exception handling**

```python
iterator = iter(my_list)
```

```python
## iterator through all elements of iterator

#next(iterator)      # lazy loading - give one element at a time
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("There is no element left in iterator")
```

### Generator<br>
Simpler way to create iterators. Previously we created iterator using list that iterator was stored in memory. But in generator using yield keyword, we can produce a series of values on the fly and do not store them in memory

when a function has yield it becomes generator object

```python
def square(n):
    for i in range(3):
        return i**2

square(3)
```

**creating generator object using yield**

```python
def square(n):
    for i in range(3):
        yield i**2
```

```python
generator_obj = square(3)       # this function call creates a generator object
```

```python
# way1 - we can run again and again
next(generator_obj)
```

```python
# way 2:
for i in generator_obj:
    print(i)
```

**For reading and working on large data file**<br>
reducing memory by not storing it, bubt directly printing or working on it.

```python
# without generator - storing in list then returning, lot of RAM use

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        lines = []
        for line in file:
            lines.append(line)
        return lines
print(read_large_file('large_file.txt'))
```

```python
def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```

```python
file_path = 'large_file.txt'

for line in read_large_file(file_path):
    print(line.strip())
```
