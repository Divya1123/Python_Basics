## Classes

```python
class Cookie:
    def __init__(self, color):
        self.color = color
    
    def get_color(self):
        return self.color
    
    def set_color(self,color):
        self.color = color
```

```python
# Creating object/instance of the class
cookie1 = Cookie('blue')
cookie2 = Cookie('pink')

print('Cookie one is:', cookie1.get_color())
print('Cookie two is:', cookie2.get_color())
```

```python
# change cookie two color
cookie2.set_color('green')

print('Cookie one is:', cookie1.get_color())
print('Cookie two is:', cookie2.get_color())
```
