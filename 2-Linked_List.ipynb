## Linked_List

### Class for creating empty node
- In most of the linked list operations like Append (i.e. adding new node at end), Prepend (i.e. adding new node at start), Insert (i.e. adding new node at specific location), we have to first create a node.
- To avoid re-writing a code we create a Node Class. Whenever we create object of Node class- a new node is created. 

```python
class Node:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = None
```

### Class for creating linked list

```python
class LinkedList:
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def print_list(self):
        temp = self.head
        while temp:
            print(temp.val)
            temp = temp.next
    
    # append item at end of linked list
    def append(self, value):
        new_node = Node(value)
        # if list is empty
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length +=1
        return True
    
    # pop item from end and return removed Node
    def pop(self):
        if self.length == 0:     # we can give self.head is None
            return None
        
        removed_node = self.tail

        if self.length == 1:
            self.head = None
            self.tail = None

        else:
            curr = self.head
            while curr.next.next:
                curr = curr.next
            curr.next = self.tail
            self.tail = curr
        self.length -=1
            
        return removed_node

    # add new node at beginning
    def prepend(self, value):
        new_node = Node(value)

        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        
        else:
            new_node.next = self.head
            self.head = new_node
        self.length += 1
        return True
    
    # remove first node
    def pop_first(self):
        if self.length == 0:
            return None
        
        temp = self.head
        self.head = self.head.next
        temp.next = None  # remove 
        
        if self.length == 0:
            self.tail = None

        self.length -= 1

    # return node at a given index
    def get(self, index):
        if index < 0 or index >= self.length:
            return None
        
        curr = self.head
        while index:
            curr = curr.next
            index -= 1
        return curr
    
    # update the value of node for given index
    def set_value(self, index, value):
        temp = self.get(index)

        if temp:
            temp.val = value
            return True
        else:
            return False
    
    # insert new node
    def insert(self, index, value):
        if index < 0 or index > self.length:
            return False
        if index == 0:
            return self.prepend(value)
        
        if index == self.length:
            return self.append(value)
        
        new_node = Node(value)
        curr = self.get(index-1)
        
        new_node.next = curr
        curr.next = new_node

        self.length +=1
        return True
    
    # remove node
    def remove(self, index):
        if index < 0 or index >= self.length:
            return None
        
        if index == 0:
            return self.pop_first()
        
        if index == self.length-1:
            return self.pop()
        
        curr = self.get(index-1)
        temp = curr.next
        curr.next = temp.next
        temp.next = None

        self.length -= 1
        return 
    
    # reverse the linked list
    def reverse(self):
        prev = None
        curr = self.head

        while curr:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr = curr.next
        return prev
        
```
