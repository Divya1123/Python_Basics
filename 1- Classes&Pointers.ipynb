### Classes

```python
class Cookie:
    def __init__(self, color):
        self.color = color
    
    def get_color(self):
        return self.color
    
    def set_color(self,color):
        self.color = color
```

```python
# Creating object/instance of the class
cookie1 = Cookie('blue')
cookie2 = Cookie('pink')

print('Cookie one is:', cookie1.get_color())
print('Cookie two is:', cookie2.get_color())
```

```python
# change cookie two color
cookie2.set_color('green')

print('Cookie one is:', cookie1.get_color())
print('Cookie two is:', cookie2.get_color())
```

### Pointers
Integers are immutable - so update by pointer, will create new int instead of updating original integer.

```python
num1 = 12
num2 = num1

print('num1:', num1)
print('num2:', num2)
print('\nnum1 points to:',id(num1))
print('num2 points to:', id(num2))
```

```python
num2 = 22

print('num1:', num1)
print('num2:', num2)
print('\nnum1 points to:',id(num1))
print('num2 points to:', id(num2))
```

Dictionary are mutable, so update via pointer will update the original memory val, instead of creating new memory

```python
dict1 = {'value':12}
dict2 = dict1

print('dict1:', dict1)
print('dict2:', dict2)
print('\ndict1 points to:',id(dict1))
print('dict2 points to:', id(dict2))
```

```python
dict2['value'] = 22

print('dict1:', dict1)
print('dict2:', dict2)
print('\ndict1 points to:',id(dict1))
print('dict2 points to:', id(dict2))
```
